### ðŸ”¹ 4. Count Vowels and Consonants

Problem:
Take a string input and count how many vowels and consonants it has.

Input:
Abhimanyu
Output:
Vowels = 4, Consonants = 5

```python
user=input("Enter a sentence:")
vowels= "aeiouAEIOU"
consonant = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"

vowels_count=0
consonant_count=0

for ch in user:
    if ch in vowels:
        vowels_count +=1
    elif ch in consonant:
        consonant_count +=1

print(f" Total vowels are: {vowels_count} , and consonant count is {consonant_count}")
```

### ðŸ”¹ 16. Count Occurrence of Each Character

Input:
"banana"
Output:
b:1 a:3 n:2

```python
user = input("Enter the Word:")
letter_count = {}  # first get the input as in dictionary so that we can check for each letter for keys and count
for n in user:
    if n in letter_count:
        letter_count[n] += 1 # if its already there add one to that letter 
    else:
        letter_count[n] = 1 # if seeeeing first time count 1 
        
print("letter occurances:")
for letter , count in letter_count.items() :
    print(letter, ":",count)
```

### ðŸ”¹ 20. Count Uppercase and Lowercase Letters

Input:
"AbHiMaNyU"

Uppercase = 5, Lowercase = 3

```python
user = input("Enter the Word:")
count_upper = 0
count_lower= 0
for ch in user:
    
    if ch.isupper():
        count_upper += 1
    elif ch.islower():
        count_lower += 1
print(f" Uppercase: {count_upper} lowercare: {count_lower}")
```
