### ‚öôÔ∏è Lesson 4: Operators in Python

üß† What is an Operator?

An operator is a symbol that tells Python to perform a specific operation (like addition, comparison, or logic).

##### üî¢ 1Ô∏è‚É£ Arithmetic Operators

Used for mathematical calculations.


| Operator | Example  | Meaning                | Output |
| -------- | -------- | ---------------------- | ------ |
| `+`      | `5 + 3`  | Addition               | 8      |
| `-`      | `5 - 3`  | Subtraction            | 2      |
| `*`      | `5 * 3`  | Multiplication         | 15     |
| `/`      | `5 / 2`  | Division (float)       | 2.5    |
| `//`     | `5 // 2` | Floor Division         | 2      |
| `%`      | `5 % 2`  | Modulus (Remainder)    | 1      |
| `**`     | `2 ** 3` | Exponentiation (Power) | 8      |


```python
x= 10
Y= 5

print(x+Y)
print(x-Y)
print(x*Y)
print(x/Y)
print(x//Y)
print(x%Y)
print(x**Y)
```

#### ‚öñÔ∏è 2Ô∏è‚É£ Comparison (Relational) Operators

Used to compare values.
They return True or False.

| Operator | Example  | Meaning          | Output |
| -------- | -------- | ---------------- | ------ |
| `==`     | `5 == 5` | Equal to         | True   |
| `!=`     | `5 != 3` | Not equal to     | True   |
| `>`      | `5 > 3`  | Greater than     | True   |
| `<`      | `5 < 3`  | Less than        | False  |
| `>=`     | `5 >= 5` | Greater or equal | True   |
| `<=`     | `5 <= 4` | Less or equal    | False  |


```python
x= 10
Y= 5

print(x==Y)
print(x!=Y)
print(x<Y)
print(x>Y)
print(x<=Y)
print(x>=Y)
```

| -------- | -------- | --------- |

| `+=`     | `x += 3` | x = x + 3 |

#### ü§î 4Ô∏è‚É£ Logical Operators

Used to combine conditional statements.

| Operator | Example              | Meaning               | Output |
| -------- | -------------------- | --------------------- | ------ |

```python
i>5
while i>0:
    print(i)
    i-=1
```

### üß© Lesson 7: Break, Continue, and Pass Statements

üß† What Are They?

They are called Loop Control Statements ‚Äî they help us control the flow of loops (either stop early, skip something, or do nothing).

| Statement  | Purpose                                           |
| ---------- | ------------------------------------------------- |
| `break`    | Stop the loop immediately                         |
| `continue` | Skip the current iteration and go to the next one |
| `pass`     | Do nothing (placeholder for future code)          |


#### üîπ 1Ô∏è‚É£ break Statement

Used to exit the loop when a certain condition is met.

```python
for i in range ( 1, 10):
    if i == 5:
        break
    print(i)
```

#### üß© Example 2: break in a while Loop

```python
i=1
while i<=10:
    print(i)
    if i==5:
        break
    i += 1
```

```python
for i in range (1,10):
    if i==5:
        continue
        
    print(i)
```

#### üß© Example 2: Print Only Even Numbers

```python
for i in range (1,10):
    if i%2 !=0:
        continue
    print(i)
```

### üîπ 3Ô∏è‚É£ pass Statement

Used when you want to do nothing ‚Äî just keep a placeholder for future code.

üß† Why use pass?
Let‚Äôs say you‚Äôre planning to write a function or loop later but don‚Äôt want an error now ‚Äî pass keeps your code valid.

```python
x = 10

if x > 5:
    pass  # will add code later
else:
    print("x is smaller than 5")
```

### üß© Lesson 8: Functions in Python

üß† What is a Function?

A function is a block of reusable code that performs a specific task.

#### üîπ 6Ô∏è‚É£ Calling One Function Inside Another

### üß© Lesson 10: Sets in Python

üîπ What is a Set?

A set in Python is a collection of unique and unordered items.
It means:

Duplicates are not allowed

Items have no specific order

It can store different types of data (int, string, etc.)

| Feature       | Description                              |
| ------------- | ---------------------------------------- |
| **Unordered** | The order of elements is not fixed       |
| **Unique**    | No duplicates allowed                    |
| **Mutable**   | You can add or remove items              |
| **Unindexed** | You cannot access elements using indexes |


```python
set={1,2,3,4}
print(set)
```

#### Basic Set Operations
1Ô∏è‚É£ Add an Element

```python
set.add(4)
print(set)
```

#### 2Ô∏è‚É£ Add Multiple Elements

```python
set.update([5,6,7])
print(set)
```

#### 3Ô∏è‚É£ Remove Elements

```python
set.remove(4)
set.discard(7)
print(set)
```

#### 4Ô∏è‚É£ Clear or Delete

```python
set.clear()
```

#### Set Operations (Math Style)

Python sets support union, intersection, difference, and symmetric difference.

‚úÖ Example 1: Union (A ‚à™ B)

Returns all unique items from both sets.

```python
set_a={1,2,3}
set_b={3,4,5}
print(set_a.union(set_b))
```

#### ‚úÖ Example 2: Intersection (A ‚à© B)

Returns items common to both sets.

```python
print(set_a.intersection(set_b))
```

#### ‚úÖ Example 3: Difference (A - B)

Returns items in A but not in B.

```python
print(set_a.difference(set_b))
```

#### ‚úÖ Example 4: Symmetric Difference (A Œî B)

Returns items that are in A or B but not both.

```python
print(set_a.symmetric_difference(set_b))
```

#### ‚úÖ Looping Through a Dictionary

You can loop through keys, values, or both.

```python
student={"name":"Abhimanyu","age":22,"Marks":99}
for key in student:
    print(key)
for value in student.values():
    print(value)
for key,value in student.items():
    print(key, ":" ,value)
```
